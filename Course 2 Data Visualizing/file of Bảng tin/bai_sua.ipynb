```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employees = pd.read_excel('employees.xls')
```

```python
employees.info()
```

```python
employees.head()
```

```python
departments = pd.read_excel('departments.xls')
```

```python
jobs = pd.read_excel('jobs.xls')
```

```python
departments
```

```python
jobs
```

```python
emp = employees[['EMPLOYEE_ID','FIRST_NAME','LAST_NAME','HIRE_DATE','SALARY','DEPARTMENT_ID','JOB_ID']]
```

```python
emp.head()
```

```python
emp = pd.merge(left=emp, right=departments[['DEPARTMENT_ID','DEPARTMENT_NAME']], left_on='DEPARTMENT_ID', 
               right_on='DEPARTMENT_ID', how='left')
```

```python
emp.head()
```

```python
emp.shape[0]
```

```python
emp[emp['DEPARTMENT_ID'].isnull()]
```

```python
emp = pd.merge(left=emp, right=jobs[['JOB_ID','JOB_TITLE']], left_on='JOB_ID', 
               right_on='JOB_ID', how='left')
```

```python
emp.head()
```

```python
emp[(emp['DEPARTMENT_ID']==50) | (emp['DEPARTMENT_ID']==90)].sort_values(by=['DEPARTMENT_ID','SALARY'], ascending=[False,True])
```

```python
emp[emp['DEPARTMENT_ID'].isin([50,90])].sort_values(by=['DEPARTMENT_ID','SALARY'], ascending=[False,True])
```

```python
emp.groupby(['DEPARTMENT_ID','DEPARTMENT_NAME']).SALARY.sum().sort_values(ascending=False).head(1).reset_index()
```

```python
df = emp.groupby(['DEPARTMENT_ID','DEPARTMENT_NAME']).SALARY.sum().to_frame().reset_index()
```

```python
df
```

```python
max_sum_sal = df.SALARY.max()
max_sum_sal
```

```python
df[df.SALARY==max_sum_sal]
```

```python
df = emp.groupby('DEPARTMENT_ID').SALARY.max().to_frame().reset_index().rename(columns={'SALARY':'MAX_SALARY'})
```

```python
df
```

```python
df_ = pd.merge(left=emp[['DEPARTMENT_ID','EMPLOYEE_ID','FIRST_NAME','SALARY']], right=df, left_on='DEPARTMENT_ID', right_on='DEPARTMENT_ID')
```

```python
df_.head()
```

```python
df_[df_['SALARY']==df_['MAX_SALARY']].sort_values(by='DEPARTMENT_ID').set_index('DEPARTMENT_ID')
```

```python
emp.groupby('DEPARTMENT_ID').SALARY.agg(['count','max','min','mean','sum'])
```

```python
emp.groupby('DEPARTMENT_NAME').size()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(data=emp, x='DEPARTMENT_NAME')
plt.title('Count of Employees per Deparment Name')
plt.ylabel('Count of Employees')
plt.xlabel('Department Name')
plt.xticks(rotation=60)
plt.grid()
```
